```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('amz_uk_price_prediction_dataset.csv')
```

```python
display(df)
```

```python
#Frequency Tables:
category_counts = df['category'].value_counts()
```

```python
#Generate a frequency table for the product category.
sns.barplot(x=df.index, y=df["category"], palette="Set3");
```

```python
top_5_categories = df['category'].value_counts().nlargest(5).index

# Filter the DataFrame to include only the top 5 categories
top_5_df = df[df['category'].isin(top_5_categories)]

# Plot the bar plot
sns.barplot(x=top_5_df.index, y=top_5_df["category"], palette="Set3")

# Show the plot
plt.show()
```

```python
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]

print(f"Mean price: {mean_price}")
print(f"Median price: {median_price}")
print(f"Mode price: {mode_price}")

# Measures of Dispersion
variance_price = df['price'].var()
std_dev_price = df['price'].std()
range_price = df['price'].max() - df['price'].min()
q1 = df['price'].quantile(0.25)
q3 = df['price'].quantile(0.75)
iqr_price = q3 - q1

print(f"Variance of price: {variance_price}")
print(f"Standard Deviation of price: {std_dev_price}")
print(f"Range of price: {range_price}")
print(f"Interquartile Range (IQR) of price: {iqr_price}")

# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True, bins=10, palette="Set3")
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['price'], palette="Set3")
plt.title('Box Plot of Product Prices')
plt.xlabel('Price')
plt.show()
```

```python
# Measures of Centrality
mean_stars = df['stars'].mean()
median_stars = df['stars'].median()
mode_stars = df['stars'].mode()[0]

print(f"Mean stars: {mean_stars}")
print(f"Median stars: {median_stars}")
print(f"Mode stars: {mode_stars}")

# Measures of Dispersion
variance_stars = df['stars'].var()
std_dev_stars = df['stars'].std()
q1 = df['stars'].quantile(0.25)
q3 = df['stars'].quantile(0.75)
iqr_stars = q3 - q1

print(f"Variance of stars: {variance_stars}")
print(f"Standard Deviation of stars: {std_dev_stars}")
print(f"Interquartile Range (IQR) of stars: {iqr_stars}")

# Shape of the Distribution
skewness_stars = df['stars'].skew()
kurtosis_stars = df['stars'].kurtosis()

print(f"Skewness of stars: {skewness_stars}")
print(f"Kurtosis of stars: {kurtosis_stars}")

# Visualizations

# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['stars'], kde=True, bins=10, palette="Set3")
plt.title('Distribution of Product Ratings (Stars)')
plt.xlabel('Stars')
plt.ylabel('Frequency')
plt.show()

# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['stars'], palette="Set3")
plt.title('Box Plot of Product Ratings (Stars)')
plt.xlabel('Stars')
plt.show()
```
